```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import urllib.request
import tensorflow as tf
from sklearn.datasets import fetch_olivetti_faces
from tensorflow.keras import datasets,layers,models
from sklearn.model_selection import train_test_split
```

```python
faces_data = fetch_olivetti_faces()
X_data = faces_data.images
Y_data = faces_data.target
print(X_data.shape)
print(Y_data.shape)

```

```python
X_resized = []
for img in X_data:
    resized_img = cv2.resize(img, (48, 48))
    X_resized.append(resized_img)

X_data = np.array(X_resized).reshape(-1, 48, 48, 1)
print(X_data.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X_data,
    Y_data,
    test_size=0.2,
    stratify=Y_data,
    random_state=41
    )
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
number_of_classes=len(np.unique(Y_data))
print(number_of_classes)
```

```python
model=models.Sequential([
    layers.RandomFlip("horizontal", input_shape=(48,48,1)),
    layers.RandomRotation(0.1),

    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(number_of_classes,activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history_2 = model.fit(X_train,Y_train, epochs=50, validation_data=(X_test,Y_test))
```

```python
test_img=X_test[7]
label=Y_test[7]

input_img=np.expand_dims(test_img, axis=0)

```

```python
predictions = model.predict(input_img)
predicted_label = np.argmax(predictions)
confidence = np.max(predictions) * 100
```

```python
plt.figure(figsize=(4, 4))
plt.imshow(test_img.reshape(48, 48), cmap='gray')
plt.title(f"Actual: {label} \n Predicted: {predicted_label} ({confidence:.2f}%)")
plt.axis('off')
plt.show()
```
